```python
df_reg = pd.read_csv(csv_file, encoding='latin1')
```

```python
df_reg.columns
```

```python
df_reg['Order Date'] = pd.to_datetime(df_reg['Order Date'])
df_reg['Ship Date'] = pd.to_datetime(df_reg['Ship Date'])
```

```python
df_reg['day_of_week'] = df_reg['Order Date'].dt.day_name()
df_reg['is_weekend'] = df_reg['day_of_week'].isin(['Tuesday', 'Sunday']).astype(int)
```

```python
df_reg['OrderY'] = df_reg['Order Date'].dt.year
df_reg['OrderM'] = df_reg['Order Date'].dt.month
```

```python
df_reg['Discounted_Sales'] = df_reg['Sales'] * (1 - df_reg['Discount'])
df_reg['Sales_per_quantity'] = df_reg['Sales']/df_reg['Quantity']
df_reg['Profit_Margin'] = df_reg['Profit']/df_reg['Sales']
df_reg['shipping_time'] = (df_reg['Ship Date'] - df_reg['Order Date']).dt.days
low, high = df_reg['Profit'].quantile([0.01, 0.99])
df_reg['profit_winsorized'] = df_reg['Profit'].clip(lower=low, upper=high)
```

```python
df_reg.columns
```

```python
numeric_df_reg = df_reg.select_dtypes(include=['int64', 'float64'])
corr = numeric_df_reg.corr()
plt.figure(figsize=(12,8))
sns.heatmap(corr, linewidths=0.25, annot=True, cmap='coolwarm')
```

```python
df_reg.columns
```

```python
df_reg.drop(columns=['Row ID', 'Order ID', 'Order Date', 'Ship Date','Customer ID',
                      'Customer Name','Country', 'City', 'State',
                      'Postal Code', 'Product ID', 'Product Name', 'Discount',
                      'day_of_week', 'is_weekend', 'shipping_time','Profit_Margin'], inplace=True)
```

```python
df_reg.info()
```

```python
X = df_reg.loc[:, df_reg.columns != 'Sales']
y = df_reg['Sales']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
```

```python
cat_features = [features for features in X_train.columns if X_train[features].dtype == 'O']
num_features = [features for features in X_train.columns if X_train[features].dtype != 'O']

preprocessor = ColumnTransformer(
    transformers = [
        ('cat', OneHotEncoder(drop='first', handle_unknown='ignore'), cat_features),
        ('num', StandardScaler(), num_features)
    ]
)

X_train_processed = preprocessor.fit_transform(X_train)
X_test_processed = preprocessor.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression, Lasso, Ridge, ElasticNet
from sklearn.ensemble import AdaBoostRegressor, RandomForestRegressor, GradientBoostingRegressor, RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import r2_score, root_mean_squared_error, mean_absolute_error, mean_squared_error
```

```python
models = {
    'LinearRegression':LinearRegression(),
    'Lasso' : Lasso(),
    'Ridge' : Ridge(),
    'ElasticNet': ElasticNet(),
    'AdaBoostRegressor' : AdaBoostRegressor(),
    'GradientBoostingRegressor' : GradientBoostingRegressor(),
    'DecisionTreeRegressor': DecisionTreeRegressor(),
    'RandomForestRegressor':RandomForestRegressor()
}
```

```python
X_train_processed.shape
```

```python
for i in range(len(list(models))):
    model = list(models.values())[i]

    # Model Training
    model.fit(X_train_processed, y_train)

    # Predictions
    y_train_pred = model.predict(X_train_processed)
    y_test_pred = model.predict(X_test_processed)

    # Shape and Size
    n_train = X_train_processed.shape[0]
    p_train = X_train_processed.shape[1]
    n_test = X_test_processed.shape[0]
    p_test = X_test_processed.shape[1]

    # Training Set Performance
    rmse_train = root_mean_squared_error(y_train, y_train_pred)
    mse_train = mean_squared_error(y_train, y_train_pred)
    mae_train = mean_absolute_error(y_train, y_train_pred)
    r2_train = r2_score(y_train, y_train_pred)
    adjusted_r2_train = 1 - (1 - r2_train) * (n_train - 1) / (n_train - p_train - 1)
    
    # Testing Set Performance
    rmse_test = root_mean_squared_error(y_test, y_test_pred)
    mse_test = mean_squared_error(y_test, y_test_pred)
    mae_test = mean_absolute_error(y_test, y_test_pred)
    r2_test = r2_score(y_test, y_test_pred)
    adjusted_r2_test = 1 - (1 - r2_test) * (n_test - 1) / (n_test - p_test - 1)

    print(list(models.keys())[i])
    
    print('Model performance for Training set')
    print("- Root Mean Squared Error: {:.4f}".format(rmse_train))
    print('- Mean Squared Error: {:.4f}'.format(mse_train))
    print('- Mean Absolute Error: {:.4f}'.format(mae_train))
    print('- R2 Score: {:.4f}'.format(r2_train))
    print('- Adjusted R2 Score: {:.4f}'.format(adjusted_r2_train))

    
    
    print('----------------------------------')
    
        
    print('Model performance for Training set')
    print("- Root Mean Squared Error: {:.4f}".format(rmse_test))
    print('- Mean Squared Error: {:.4f}'.format(mse_test))
    print('- Mean Absolute Error: {:.4f}'.format(mae_test))
    print('- R2 Score: {:.4f}'.format(r2_test))
    print('- Adjusted R2 Score: {:.4f}'.format(adjusted_r2_test))

    
    print('='*35)
    print('\n')


```

```python
from sklearn.linear_model import RidgeCV

ridge_cv = RidgeCV(alphas=[0.01, 0.1, 1, 10, 100], cv=5)
ridge_cv.fit(X_train_processed, y_train)

print("Best alpha:", ridge_cv.alpha_)
```
